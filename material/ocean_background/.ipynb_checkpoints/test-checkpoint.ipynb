```python
angle = 360

k1 = (360-angle)/360
k2 = 1-k1

z = 10*np.sin(k1*x/5 + k2*y/5 + 15)

plt.imshow(z)
```

```python
waves = 10
height = 5

def z_waves(x,y,angle,height,waves,phase):
    k1 = (360-angle)/360
    k2 = 1-k1
    return height*np.sin(k1*x/waves + k2*y/waves + phase)

angle = 45
z = z_waves(x,y,angle)
for a in np.random.randint(0,360,5):
    z += z_waves(x,y,a)

plt.imshow(z)
```
