```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pixel_x = 768
pixel_y = 824

angle = 0

```

```python
np.zeros((pixel_x,pixel_y))
```

```python
x,y = np.meshgrid(np.arange(pixel_y),np.arange(pixel_x))
```

```python
angle = 360

k1 = (360-angle)/360
k2 = 1-k1

z = 10*np.sin(k1*x/5 + k2*y/5)+40*np.random.random((pixel_x,pixel_y))

plt.imshow(z)
```

```python
angle = 360

k1 = (360-angle)/360
k2 = 1-k1

z = 5*np.sin(k1*x/30 + k2*y/30 + 15)

plt.imshow(z)
```

```python
waves = 30
height = 10

def z_waves(x,y,angle,height,waves,phase):
    k1 = (360-angle)/360
    k2 = 1-k1
    return height*np.sin(k1*x/waves + k2*y/waves + phase)

angle = 45
z = z_waves(x,y,angle,waves,height,0)

for a in np.random.randint(0,360,5):
    z += z_waves(x,y,a,waves,height,0)

plt.imshow(z,cmap='Blues',vmin=-600)
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()

wave_img = plt.pcolormesh(z, cmap='Blues', vmin=0)

def z_wave2(phase):
    return wave_img.set_array( 10*np.sin(k1*x/5 + k2*y/5 + phase).ravel() )

myAnimation = animation.FuncAnimation(fig, z_wave2, frames=np.arange(0,200,1), interval=100, blit=False, repeat=True)
myAnimation
```
